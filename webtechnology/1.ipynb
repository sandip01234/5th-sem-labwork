```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
```

```python
# from google.colab import files
# uploaded = files.upload()  # Upload your darkened image here

# # Load image in grayscale
# img = cv2.imread(list(uploaded.keys())[0], cv2.IMREAD_GRAYSCALE)
img= cv2.imread('/content/hist_equal_inputimg.PNG',0)

# Histogram Equalization
equalized_img = cv2.equalizeHist(img)

# Show original and equalized images
plt.subplot(1, 2, 1)
plt.title('Original Dark Image')
plt.imshow(img, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('After Histogram Equalization')
plt.imshow(equalized_img, cmap='gray')

plt.show()
```

Advantages of Histogram Equalization:

Enhances contrast by redistributing intensity values.

Improves visibility of details in dark or low-contrast images.

Works well for images with uneven lighting.

Fully automated, requiring no manual parameter tuning.
