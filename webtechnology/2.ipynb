### Noise
Noise in an image refers to unwanted random variations in brightness or color information. It distorts the original image data, making it difficult to process or analyze.

### Salt and Pepper Noise:
Salt and pepper noise is a type of impulse noise.

It appears as random occurrences of black (pepper) and white (salt) pixels scattered across the image.

It mainly occurs due to errors in image transmission or faulty sensors.

### Effect of Noise on Image Processing
1. Reduces image quality and clarity.

2. Affects edge detection and object recognition accuracy.

3. Leads to loss of important details in images.

4. Introduces artificial pixels that mislead algorithms.

### Technique used: Median Filtering

It replaces each pixel value with the median of its neighboring pixel values.

Best for removing salt and pepper noise because it preserves edges while eliminating isolated noisy pixels.


```python
# Import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# # Upload the image
# from google.colab import files
# uploaded = files.upload()

# # Read the uploaded image in grayscale
# image_path = list(uploaded.keys())[0]
# img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

img= cv2.imread('/content/drive/MyDrive/Colab Notebooks/imageprocessingimages/Q2_2_salt andpepper img _input img.tif',0)

# Apply Median Filter to remove Salt and Pepper Noise
denoised_img = cv2.medianBlur(img, 3)

# Display the denoised image
plt.imshow(denoised_img, cmap='gray')
plt.title('Denoised Image (Median Filter)')
plt.axis('off')
plt.show()

```

### Edge Detection

Edge detection is the process of identifying and locating sharp discontinuities (edges) in an image.

###Filter used: Canny Edge Detection

Canny is a multi-stage algorithm detecting a wide range of edges in images.

It first reduces noise, finds gradients, applies non-maximum suppression, and finally uses double thresholding to detect edges.

###Significance of Canny Edge Detector:

Highly accurate in detecting thin, clean edges.

Reduces the effect of noise because of built-in Gaussian filtering.

Detects both strong and weak edges.

```python
# Apply Canny Edge Detection
edges = cv2.Canny(denoised_img, 100, 200)

# Display the edge-detected image
plt.imshow(edges, cmap='gray')
plt.title('Edges Detected (Canny)')
plt.axis('off')
plt.show()
```
